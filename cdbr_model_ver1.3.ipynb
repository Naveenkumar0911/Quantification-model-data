```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
sc=StandardScaler()
lc=LabelEncoder()

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
```

```python
cr_df=pd.read_csv('_7CR_50Datapoints.csv')
cdbr_df=pd.read_csv('cdbr_50random_Dataset.csv')
```

```python
cr_df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
cr_df.head(5)
```

```python
cdbr_df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)

cdbr_df.head(15)
```

```python
cdbr_df=cdbr_df[cdbr_df.CR_Categories!='Temp_category']
```

```python
print(cdbr_df.shape)

print(cr_df.shape)
```

```python
cdbr_random=pd.read_csv('cdbr_randomdata.csv')
cdbr_random.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
cdbr_random=cdbr_random[cdbr_random.CR_Categories!='Temp_category']
cdbr_random.head(14)
```

```python
final=pd.DataFrame()
for data in cdbr_random['Datapoint'].unique():
    temp=cdbr_random[cdbr_random['Datapoint']==data]
    temp=temp.drop(columns=['Datapoint'])
    temp=temp.set_index(['CR_Categories'])
    temp=temp.transpose()
    temp['Datapoint']=data
    final=pd.concat((final,temp))
    
    
    
for i in range(2,len(cr_df.columns)):
    col=cr_df.columns[i]
    temp=list(cr_df[col])
    final[col]=temp
final
```

```python
x=final.reset_index()
x=x.rename(columns={'index':'CR_Categories'})
x['CR_Categories']=lc.fit_transform(x['CR_Categories'])
x
```

```python
x_ID = x.loc[x["Datapoint"]=="Datapoint0"]
x_ID.drop(labels=['Datapoint'],axis=1,inplace=True)
x_final = np.expand_dims(x_ID.to_numpy().flatten(),axis=0)

for data in x['Datapoint'].unique()[1:]:
    x_ID = x.loc[x["Datapoint"]==data]
    x_ID.drop(labels=['Datapoint'],axis=1,inplace=True)
    x_final =np.vstack((x_final,np.expand_dims(x_ID.to_numpy().flatten(),axis=0)))
```

```python
x_final.shape
```

```python
y=cdbr_df.copy()
y['CR_Categories']=lc.fit_transform(y['CR_Categories'])
y
```

```python
y_ID = y.loc[y["Datapoint"]=="Datapoint0"]
y_ID.drop(labels=['Datapoint'],axis=1,inplace=True)
y_new = np.expand_dims(y_ID.to_numpy().flatten(),axis=0)

for data in y['Datapoint'].unique()[1:]:
    y_ID = y.loc[y["Datapoint"]==data]
    y_ID.drop(labels=['Datapoint'],axis=1,inplace=True)
    y_new =np.vstack((y_new,np.expand_dims(y_ID.to_numpy().flatten(),axis=0)))
```

```python
print(x_final.shape)
print(y_new.shape)
```

```python
train_x,test_x,train_y,test_y=train_test_split(x_final,y_new,test_size=0.2,random_state=2)
```

```python
from xgboost import XGBRegressor

model=XGBRegressor(max_depth=5,reg_alpha=1,reg_lambda=1,
                   colsample_bytree=0.7,n_estimators=1000,
                   learning_rate=0.001,objective='reg:squarederror')

model.fit(train_x,train_y)
```

```python
pred_y=model.predict(test_x)

pred_train_y=model.predict(train_x)

rmse_test = np.sqrt(mse(test_y, pred_y))

rmse_train = np.sqrt(mse(train_y, pred_train_y))

print("RMSE Test : % f" %(rmse_test))

print("RMSE Train: % f" %(rmse_train))
```
