## Severity Calculation for Cyber Risk Model

```python
cr_severity=cr_random_data.copy()
```

```python
unique_datapoints=Sectorwise_consequences['Datapoint'].unique()
unique_values=Sectorwise_consequences['Consequences'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=Sectorwise_consequences[Sectorwise_consequences['Datapoint']==data].loc[:,'Severity'].iloc[i]
        s_con_val=Sectorwise_consequences[Sectorwise_consequences['Datapoint']==data].loc[:,'Consequences'].iloc[i]
        
        
        for j in range(0,13):
            
            cr_cat=cr_severity['CR_Categories'].iloc[j]
            val=cr_severity[cr_severity['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con
            
            filt=(cr_severity['Datapoint']==data) & (cr_severity['CR_Categories']==cr_cat)
            
            cr_severity.loc[filt,s_con_val]=cr_loc
```

```python
cr_severity['Total']=cr_severity.max(axis=1)
cr_severity
```

## Likelihood Calculation for Cyber Risk Model

```python
cr_likelihood=cr_random_data.copy()

unique_datapoints=Sectorwise_consequences['Datapoint'].unique()
unique_values=Sectorwise_consequences['Consequences'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=Sectorwise_consequences[Sectorwise_consequences['Datapoint']==data].loc[:,'Likelihood'].iloc[i]
        s_con_val=Sectorwise_consequences[Sectorwise_consequences['Datapoint']==data].loc[:,'Consequences'].iloc[i]
        
        
        for j in range(0,13):
            
            cr_cat=cr_likelihood['CR_Categories'].iloc[j]
            val=cr_likelihood[cr_likelihood['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con
            
            filt=(cr_likelihood['Datapoint']==data) & (cr_likelihood['CR_Categories']==cr_cat)
            
            cr_likelihood.loc[filt,s_con_val]=cr_loc
```

####  likelihood max average total calculation

```python
cr_likelihood['Max_avg_Total']=np.zeros((len(cr_likelihood),1))
for z in range(0,len(cr_likelihood)):    
    list1=[]
    for i in range(2,len(cr_likelihood.columns)):
        col=cr_likelihood.columns[i]
        list1.append(cr_likelihood.loc[z,col])

    if list1[0]>list1[1]:
        largest=list1[0]
        slarget=list1[1]
    else:
        largest=list1[1]
        slarget=list1[0]

    for j in range(2,len(list1)):
        if list1[j]>largest:
            slarget=largest
            largest=list1[j]
        elif list1[j]>slarget and list1[j]<=largest:
            slarget=list1[j]
    avg=(largest+slarget)/2
    cr_likelihood['Max_avg_Total'].iloc[z]=avg
    #print(largest,slarget)
```

## Remediation Cost Calculation for Cyber Risk Model

##### calculating sum of CR_Random data for every sets

```python
y=pd.DataFrame()

col_name=1

for i in range(0,len(cr_random_data_new)-1,13):
    
    j=i+13
    
    x=cr_random_data_new.iloc()[i:j].sum()
    
    y=pd.concat((y,x),axis=1)
    
    y=y.rename(columns={0:col_name})
    
    col_name+=1
```

```python
y=y.transpose()
cr_random_sum=y.reindex(columns=['Ransom','Service degradation','IP Theft','Reputation hit','Low quality Product','Data Loss','Privacy Breach','OT Failure','Resource Theft'])
cr_random_sum
```

```python
Sectorwise_consequences_new=Sectorwise_consequences.copy()
Sectorwise_consequences_new['Remediation Cost_After_Random_cal']=np.zeros((len(Sectorwise_consequences_new),1))
Sectorwise_consequences_new
```

```python
z=0
#new_rc=list()

for i in range(0,len(Sectorwise_consequences_new),9):
    z+=1
    for j in range(i,i+9):
        rc_val=Sectorwise_consequences_new['Remediation Cost'].iloc[j]
        rc_conse_type=Sectorwise_consequences_new['Consequences'].iloc[j]
        
        if z==51:
            break
        else:

            cr_random_sum_val=cr_random_sum.loc[z,rc_conse_type]
            output=rc_val/cr_random_sum_val
            #new_rc.append(output)
        Sectorwise_consequences_new['Remediation Cost_After_Random_cal'].iloc[j]=output
```

```python
Sectorwise_consequences_new
```

```python
cr_remediationcost=cr_random_data.copy()

unique_datapoints=Sectorwise_consequences_new['Datapoint'].unique()
unique_values=Sectorwise_consequences_new['Consequences'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=Sectorwise_consequences_new[Sectorwise_consequences_new['Datapoint']==data].loc[:,'Remediation Cost_After_Random_cal'].iloc[i]
        s_con_val=Sectorwise_consequences_new[Sectorwise_consequences_new['Datapoint']==data].loc[:,'Consequences'].iloc[i]

        
        for j in range(0,13):
            
            cr_cat=cr_remediationcost['CR_Categories'].iloc[j]
            val=cr_remediationcost[cr_remediationcost['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con*5
            
            filt=(cr_remediationcost['Datapoint']==data) & (cr_remediationcost['CR_Categories']==cr_cat)
            
            cr_remediationcost.loc[filt,s_con_val]=cr_loc
```
