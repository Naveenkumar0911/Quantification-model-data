## Severity Calculation for CDBR Model

```python
cdbr_severity=cdbr_random_data.copy()

unique_datapoints=cyber_risk['Datapoint'].unique()
unique_values=cyber_risk['CR_Categories'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=cyber_risk[cyber_risk['Datapoint']==data].loc[:,'cr_Severity'].iloc[i]
        s_con_val=cyber_risk[cyber_risk['Datapoint']==data].loc[:,'CR_Categories'].iloc[i]
        
        
        for j in range(0,12):
            
            cr_cat=cdbr_severity['CR_Categories'].iloc[j]
            val=cdbr_severity[cdbr_severity['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con
            

            filt=(cdbr_severity['Datapoint']==data) & (cdbr_severity['CR_Categories']==cr_cat)
            
            cdbr_severity.loc[filt,s_con_val]=cr_loc
            
```

```python
cdbr_severity['Total']=cdbr_severity.max(axis=1)
cdbr_severity
```

## Likelihood Calculation for Cyber Risk Model

```python
cdbr_likelihood=cdbr_random_data.copy()

unique_datapoints=cyber_risk['Datapoint'].unique()
unique_values=cyber_risk['CR_Categories'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=cyber_risk[cyber_risk['Datapoint']==data].loc[:,'cr_Likelihood'].iloc[i]
        s_con_val=cyber_risk[cyber_risk['Datapoint']==data].loc[:,'CR_Categories'].iloc[i]
        
        
        for j in range(0,12):
            
            cr_cat=cdbr_likelihood['CR_Categories'].iloc[j]
            val=cdbr_likelihood[cdbr_likelihood['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con
            
            filt=(cdbr_likelihood['Datapoint']==data) & (cdbr_likelihood['CR_Categories']==cr_cat)
            
            cdbr_likelihood.loc[filt,s_con_val]=cr_loc
```

##### likelihood max average total calculation

```python
cdbr_likelihood['Max_avg_Total']=np.zeros((len(cdbr_likelihood),1))
for z in range(0,len(cdbr_likelihood)):    
    list1=[]
    for i in range(2,len(cdbr_likelihood.columns)):
        col=cdbr_likelihood.columns[i]
        list1.append(cdbr_likelihood.loc[z,col])

    if list1[0]>list1[1]:
        largest=list1[0]
        slarget=list1[1]
    else:
        largest=list1[1]
        slarget=list1[0]

    for j in range(2,len(list1)):
        if list1[j]>largest:
            slarget=largest
            largest=list1[j]
        elif list1[j]>slarget and list1[j]<=largest:
            slarget=list1[j]
    avg=(largest+slarget)/2
    cdbr_likelihood['Max_avg_Total'].iloc[z]=avg
    #print(largest,slarget)
```

## Remediation Cost Calculation for Cyber Risk Model

```python
y=pd.DataFrame()

col_name=1

for i in range(0,len(cdbr_random_data_new)-1,12):
    
    j=i+12
    
    x=cdbr_random_data_new.iloc()[i:j].sum()
    
    y=pd.concat((y,x),axis=1)
    
    y=y.rename(columns={0:col_name})
    
    col_name+=1
```

```python
y=y.transpose()
cr_random_sum=y.drop(columns=(['Datapoint','CR_Categories']),axis=1)
cr_random_sum
```

```python
cyber_risk_new=cyber_risk.copy()
cyber_risk_new['Remediation Cost_After_Random_cal']=np.zeros((len(cyber_risk_new),1))
cyber_risk_new
```

```python
z=0

for i in range(0,len(cyber_risk_new),13):
    z+=1
    for j in range(i,i+13):
        rc_val=cyber_risk_new['cr_remediationcost_total'].iloc[j]
        rc_conse_type=cyber_risk_new['CR_Categories'].iloc[j]
        
        if z==51:
            break
        else:

            cr_random_sum_val=cr_random_sum.loc[z,rc_conse_type]
            output=rc_val/cr_random_sum_val
            #new_rc.append(output)
        cyber_risk_new['Remediation Cost_After_Random_cal'].iloc[j]=output
```

```python
cdbr_remediationcost=cdbr_random_data.copy()

unique_datapoints=cyber_risk_new['Datapoint'].unique()
unique_values=cyber_risk_new['CR_Categories'].unique()

for data in unique_datapoints:
    
    for i in range(0,len(unique_values)):
        
        s_con=cyber_risk_new[cyber_risk_new['Datapoint']==data].loc[:,'Remediation Cost_After_Random_cal'].iloc[i]
        s_con_val=cyber_risk_new[cyber_risk_new['Datapoint']==data].loc[:,'CR_Categories'].iloc[i]

        
        for j in range(0,12):
            
            cr_cat=cdbr_remediationcost['CR_Categories'].iloc[j]
            val=cdbr_remediationcost[cdbr_remediationcost['Datapoint']==data].loc[:,s_con_val].iloc[j]
            
            cr_loc= val*s_con
            
            filt=(cdbr_remediationcost['Datapoint']==data) & (cdbr_remediationcost['CR_Categories']==cr_cat)
            
            cdbr_remediationcost.loc[filt,s_con_val]=cr_loc
```

```python
cdbr_remediationcost['cdbr_remediationcost_total']=cdbr_remediationcost.sum(axis=1)
```

```python
cdbr_final=cdbr_remediationcost.loc[:,['Datapoint','CR_Categories','cdbr_remediationcost_total']]
cdbr_final['cdbr_Likelihood']=cdbr_likelihood.loc[:,'Max_avg_Total']
cdbr_final['cdbr_Severity']=cdbr_severity.loc[:,'Total']
cdbr_final
```
