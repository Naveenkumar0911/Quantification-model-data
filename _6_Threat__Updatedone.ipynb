```python
import pandas as pd
import numpy as np
```

## Part -1 Threat Severity Calculation

```python
### variables
cat=pd.DataFrame()

threat_cat=pd.DataFrame()

for i in range(0,50):
    
    threat_cat['threat actor']=['Organized Crime','Other Criminals','APTs','Other Governments','Competitors','Anarchist Insider','Anarchist Outsider']

    threat_cat['Technology']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Expertise']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Resources']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Access']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Generally']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Specifically']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Low-Conscious']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Perception-Conscious']=np.random.randint(1,11,size=(len(threat_cat)))

    threat_cat['Timing-Sensitivity']=np.random.randint(1,11,size=(len(threat_cat)))
    
    cat=pd.concat([threat_cat,cat])

cat['Total']=cat.sum(axis=1)

threat_severity=cat[['threat actor','Total']]
```

## Part -2 Intent Random Data Generation and map with tsc adjustment

```python
intent=pd.DataFrame()

df=pd.DataFrame()

for i in range(0,50):
    
    df['Threat Actor']=['Organized Crime','Other Criminals','APTs','Other Governments','Competitors','Anarchist Insider','Anarchist Outsider']
    
    df['Ransom']=np.random.randint(0,6,size=len(df))
    
    df['Service degradation']=np.random.randint(0,6,size=len(df))
    
    df['IP Theft']=np.random.randint(0,6,size=len(df))
    
    df['Reputation hit']=np.random.randint(0,6,size=len(df))
    
    df['Low quality Product']=np.random.randint(0,6,size=len(df))
    
    df['Data Loss']=np.random.randint(0,6,size=len(df))
    
    df['Privacy Breach']=np.random.randint(0,6,size=len(df))
    
    df['OT Failure']=np.random.randint(0,6,size=len(df))
    
    df['Resource Theft']=np.random.randint(0,6,size=len(df))
    
    intent=pd.concat([intent,df])
```

```python
intent_df=intent.copy()
intent
```

#### tsc adjustment

```python
tsc_adjustment={1:-.33,2:-.20,3:0,4:.20,5:.33,0:0}
```

```python
j=1

col=len(intent_df.columns)

for i in range(j,col):
        
    col_name=intent_df.columns[i]
    
    #new_name=intent_df.columns[i]+"_TSC_Adjustement"
    
    intent_df[col_name]=(intent_df[col_name].map(tsc_adjustment))+1
```
