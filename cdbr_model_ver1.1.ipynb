```python
import pandas as pd
import numpy as np
```

#### reading cdbr and CR datasets

```python
cdbr_df=pd.read_csv('cdbr_50random_Dataset.csv')
csr_df=pd.read_csv('_7CR_50Datapoints.csv')
```

#### Standardize both the datasets

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms=MinMaxScaler()

scaled_csr=mms.fit_transform(csr_df.loc[:,['cr_remediationcost_total','cr_Likelihood','cr_Severity']])

scaled_csr=pd.DataFrame(scaled_csr)

scaled_csr.rename(columns={0:'remediation cost',1:'Likelihood',2:'Severity'},inplace=True)

scaled_csr=pd.concat((scaled_csr,csr_df.loc[:,['Datapoint','CR_Categories']]),axis=1)

scaled_csr
```

```python
scaled_cdbr=mms.fit_transform(cdbr_df.loc[:,['cdbr_remediationcost_total','cdbr_Likelihood','cdbr_Severity']])

scaled_cdbr=pd.DataFrame(scaled_cdbr)

scaled_cdbr.rename(columns={0:'remediation cost',1:'Likelihood',2:'Severity'},inplace=True)

scaled_cdbr=pd.concat((scaled_cdbr,cdbr_df.loc[:,['Datapoint','CR_Categories']]),axis=1)

scaled_cdbr
```

## compliance failure

#### seprate compliance failure category datapoints from the Cyber Risk Dataset and create X and Y dataset for spliting

```python
x=scaled_csr[scaled_csr['CR_Categories']=='Compliance Failure']

x.drop(columns=['CR_Categories','Datapoint'],axis=1,inplace=True)

x
```

```python
cdbr_ID = scaled_cdbr.loc[scaled_cdbr["Datapoint"]=="Datapoint0"]
cdbr_ID.drop(labels=['Datapoint','CR_Categories'],axis=1,inplace=True)
y = np.expand_dims(cdbr_ID.to_numpy().flatten(),axis=0)

for data in scaled_cdbr['Datapoint'].unique()[1:]:
    cdbr_ID = scaled_cdbr.loc[scaled_cdbr["Datapoint"]==data]
    cdbr_ID.drop(labels=['Datapoint','CR_Categories'],axis=1,inplace=True)
    y =np.vstack((y,np.expand_dims(cdbr_ID.to_numpy().flatten(),axis=0)))
```

```python
y.shape
```

#### importing libraries for modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error as mse
```

```python
#### creating a random forest regressor

rfr=RandomForestRegressor(n_estimators=50,max_depth=40,random_state=0)

### splitting dataset

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.80,random_state=0)

### create model

rfr_multir=MultiOutputRegressor(rfr)

rfr_multir.fit(x_train,y_train)
```

```python
### fittin the model and checking score

y_pred_multir=rfr_multir.predict(x_test)

score=rfr_multir.score(x_train,y_train)

print('Training Score :',score)

rmse = np.sqrt(mse(y_test, y_pred_multir))

print(rmse)
```

```python
### checking training and testing dataset RMSE 

y_train_pred=rfr_multir.predict(x_train)

y_test_pred=rfr_multir.predict(x_test)

rmse_train = np.sqrt(mse(y_train, y_train_pred))

rmse_test = np.sqrt(mse(y_test, y_test_pred))


print("RMSE : % f" %(rmse_train))

print("RMSE : % f" %(rmse_test))
```

## try the model using ANN deep learning method

```python
import tensorflow
```

```python
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential

x_train1,x_test1,y_train1,y_test1=train_test_split(x,y,train_size=.80,random_state=0)
```

```python
# get the model
def get_model(n_inputs, n_outputs):
    model = Sequential()
    model.add(Dense(40, input_dim=n_inputs, kernel_initializer='he_uniform', activation='sigmoid'))
    model.add(Dropout(0.5))
    model.add(Dense(30, kernel_initializer='he_uniform', activation='sigmoid'))
    
    model.add(Dense(15,kernel_initializer='he_uniform', activation='sigmoid'))
    model.add(Dense(n_outputs, kernel_initializer='he_uniform',activation='linear'))
    model.compile(loss='mae', optimizer='adam')
    return model
```

```python
n_inputs, n_outputs = x.shape[1], y.shape[1]
```

```python
model = get_model(n_inputs, n_outputs)

model.fit(x_train1, y_train1, verbose=0, epochs=100,batch_size=10)
```

```python
yhat = model.predict(x_test1)
```

```python
rmse = np.sqrt(mse(y_test1, yhat))
rmse
```

## try on LSTM model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,LSTM
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x,y,train_size=.80,random_state=0)

print(x_train1.shape,x_test1.shape,y_train1.shape,y_test1.shape)
```

```python
x_train1=np.array(x_train1)
x_test1=np.array(x_test1)
```

```python
## reshape input which is require for LSTM

x_train1=x_train1.reshape(x_train1.shape[0],x_train1.shape[1],1)

x_test1=x_test1.reshape(x_test1.shape[0],x_test1.shape[1],1)
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(3,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(10))
model.add(Dense(36))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(x_train1,y_train1,validation_data=(x_test1,y_test1),epochs=100,batch_size=60,verbose=1)
```

```python
### prediction and chck the performance metrics

train_predict=model.predict(x_train1)

test_predict=model.predict(x_test1)
```

```python
##training rmse

import math

math.sqrt(mse(y_train1,train_predict))
```

```python
##testing rmse

math.sqrt(mse(y_test1,test_predict))
```
