```python
import pandas as pd
import numpy as np
```

```python
min_val=0
max_val=10000
revenue=1
no_of_controls=3
impact=3
likelihood=3
```

```python
df_product=pd.read_csv("Products Categories.csv")
df_control=pd.read_csv("Products Categories.csv")
```

```python
df_product
```

```python
final_console_data=pd.DataFrame()
```

```python
for i in range(0,50):
    
    ######################################################## Products ##########################################################

    df_product['weights']=50
    df_product['score']=np.random.randint(1,6,size=len(df_product))
    df_product['weighted_score']=df_product['weights']*df_product['score']

    ### Default risk exposure range based on score

    dre_1_min=min_val
    dre_2_min=max_val/40
    dre_3_min=max_val/10
    dre_4_min=max_val/4
    dre_5_min=max_val/2


    dre_1_max=dre_2_min-1
    dre_2_max=dre_3_min-1
    dre_3_max=dre_4_min-1
    dre_4_max=dre_5_min-1
    dre_5_max=max_val

    re_min_value={1:dre_1_min,2:dre_2_min,3:dre_3_min,4:dre_4_min,5:dre_5_min}

    re_max_value={1:dre_1_max,2:dre_2_max,3:dre_3_max,4:dre_4_max,5:dre_5_max}


    df_product['re_min_range']=df_product['score'].map(re_min_value)
    df_product['re_max_range']=df_product['score'].map(re_max_value)


    df_product['re-min']=(df_product['re_min_range']*impact)*revenue
    df_product['re-max']=(df_product['re_max_range']*impact)*revenue

    df_product['ale']=((df_product['re-min']+df_product['re-max'])/2)*(likelihood/5)


    ### Default risk exposure range based on score

    drc_1_min=dre_1_min*0.2
    drc_2_min=dre_2_min*0.2
    drc_3_min=dre_3_min*0.2
    drc_4_min=dre_4_min*0.2
    drc_5_min=dre_5_min*0.2


    drc_1_max=dre_1_max*0.2
    drc_2_max=dre_2_max*0.2
    drc_3_max=dre_3_max*0.2
    drc_4_max=dre_4_max*0.2
    drc_5_max=dre_5_max*0.2

    rc_min_value={1:drc_1_min,2:drc_2_min,3:drc_3_min,4:drc_4_min,5:drc_5_min}

    rc_max_value={1:drc_1_max,2:drc_2_max,3:drc_3_max,4:drc_4_max,5:drc_5_max}

    df_product['rc_min_range']=df_product['score'].map(rc_min_value)
    df_product['rc_max_range']=df_product['score'].map(rc_max_value)

    df_product['rc-min']=(df_product['rc_min_range']*likelihood)*revenue
    df_product['rc-max']=(df_product['rc_max_range']*likelihood)*revenue

    df_new=df_product.drop(columns=['re_min_range','re_max_range','rc_min_range','rc_max_range'],axis=1)


    pr_df=df_new.copy()


    #######controls#####################################################################################################



    df_control['weights']=50

    df_control['score']=np.random.randint(1,6,size=len(df_control))

    df_control['weighted_score']=df_control['weights']*df_control['score']



    df_control['score_final']=df_control.score.apply(lambda x:1 if x<3 else 5)

    ### Default risk exposure range based on score

    dre_1_min=min_val
    dre_2_min=max_val/40
    dre_3_min=max_val/10
    dre_4_min=max_val/4
    dre_5_min=max_val/2


    dre_1_max=dre_2_min-1
    dre_2_max=dre_3_min-1
    dre_3_max=dre_4_min-1
    dre_4_max=dre_5_min-1
    dre_5_max=max_val

    re_min_value={1:dre_1_min,2:dre_2_min,3:dre_3_min,4:dre_4_min,5:dre_5_min}

    re_max_value={1:dre_1_max,2:dre_2_max,3:dre_3_max,4:dre_4_max,5:dre_5_max}

    df_control['re_min_range']=df_control['score_final'].map(re_min_value)
    df_control['re_max_range']=df_control['score_final'].map(re_max_value)

    df_control['re-min']=(df_control['re_min_range']*impact)*revenue
    df_control['re-max']=(df_control['re_max_range']*impact)*revenue

    df_control['ale']=(((df_control['re-min']+df_control['re-max'])/2)*(likelihood/5))*revenue

    drc_1_min=dre_1_min*0.2
    drc_2_min=dre_2_min*0.2
    drc_3_min=dre_3_min*0.2
    drc_4_min=dre_4_min*0.2
    drc_5_min=dre_5_min*0.2


    drc_1_max=dre_1_max*0.2
    drc_2_max=dre_2_max*0.2
    drc_3_max=dre_3_max*0.2
    drc_4_max=dre_4_max*0.2
    drc_5_max=dre_5_max*0.2

    rc_min_value={1:drc_1_min,2:drc_2_min,3:drc_3_min,4:drc_4_min,5:drc_5_min}

    rc_max_value={1:drc_1_max,2:drc_2_max,3:drc_3_max,4:drc_4_max,5:drc_5_max}


    df_control['rc_min_range']=df_control['score'].map(rc_min_value)
    df_control['rc_max_range']=df_control['score'].map(rc_max_value)


    df_control['rc-min']=(df_control['rc_min_range']*likelihood)*revenue
    df_control['rc-max']=(df_control['rc_max_range']*likelihood)*revenue

    df_new=df_control.drop(columns=['re_min_range','re_max_range','rc_min_range','rc_max_range','score'],axis=1)

    ctrl_df=df_new.copy()

    #####################################final impact csv######################################################################

    ctrl_df=ctrl_df.rename(columns={'score_final':'score'})

    df_merged=pr_df.append(ctrl_df)

    cons_mod=df_merged.groupby(['Consequences'])[['weighted_score','weights','re-min','re-max','rc-min','rc-max']].sum()

    cons_mod=pd.DataFrame(cons_mod)

    cons_mod['Impact Score']=np.random.randint(1,11,size=len(cons_mod))

    cons_mod=cons_mod.rename(columns={'weighted_score':'Vulnerability Weighted Score','weights':'Vulnerability Weight'})

    cons_mod['Vulnerability Score']=cons_mod['Vulnerability Weighted Score']/cons_mod['Vulnerability Weight']
    cons_mod

    cons_mod['Vulnerability Remediation Cost']=cons_mod['rc-min']+cons_mod['rc-max']

    cons_mod['Vulnerability Exposure Cost']=cons_mod['re-min']+cons_mod['re-max']


    console_data=cons_mod.drop(columns=['re-min','re-max','rc-min','rc-max'],axis=1)
    
    console_data=console_data.reset_index(drop=False)
    
    final_console_data=pd.concat((final_console_data,console_data),axis=0)
    
    
    
```

```python
final_console_data.to_csv('Impact_random_generated_data.csv')
```
